```python
import pandas as pd
```

```python
def find_s(file_path):
    data = pd.read_csv(file_path)

    attributes = data.columns[:-1]
    target = data.columns[-1]
    hypothesis = ["?"] * len(attributes)

    for _, row in data.iterrows():
        if row[target] != "Yes":
            continue
        
        for i, value in enumerate(row[attributes]):
            if hypothesis[i] in ("?", value):
                hypothesis[i] = value
            else:
                hypothesis[i] = "?"

    print(f"{data}\n\nFinal Hypothesis: {hypothesis}")
```

```python
find_s("lab04.csv")
```
